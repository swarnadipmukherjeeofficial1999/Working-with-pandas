## Read .csv files:

## Writing to csv files :

```python
new_data=pd.DataFrame({'ROLL_NO':[5,6],
                        'NAME':['Adams','Abhishek'],
                        'SUBJECT':['cse','cse'],
                        'MARKS':[56,85]},
                         index=[17,18])

students_updated=pd.concat([students_df,new_data])
```

```python
students_updated
```

```python
students_updated.to_csv('new_student_sample.csv',index=False)
```
